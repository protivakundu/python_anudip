```python
#constant
from scipy import constants as cons
```

```python
#diractory inside constants
dir(cons)
```

```python
#matric----unit meter
print(cons.kilo)
print(cons.deci)
print(cons.centi)
```

```python
#binary prefixes----unit in bytes
print(cons.kibi)
print(cons.mebi)
print(cons.gibi)
```

```python
#mass-----unit in kilo gram
print(cons.gram)
print(cons.pound) #lb(pound)
print(cons.carat)
```

```python
#time---unit in seconds
print(cons.minute)
print(cons.hour)
print(cons.day)
print(cons.week)
print(cons.year)
```

```python
#length----unit in meter
print(cons.inch)
print(cons.foot)
print(cons.mile)
print(cons.yard)
```

```python
#tempareture----unit in Kelvin
print(cons.degree_Fahrenheit)
print(cons.zero_Celsius)
```

```python
#pressure-----unit in pascals
print(cons.atm)   #atm(atmosphere)
print(cons.psi)
```

```python
#area----unit in squre meters
print(cons.hectare)
print(cons.acre)
```

```python
#volumne-----unit in cubic meter
print(cons.litre)
print(cons.gallon)
```

```python
#speed----unit in meter per second

print(cons.kmh)
print(cons.mph)
print(cons.mach)
print(cons.speed_of_sound)
print(cons.knot)
```

```python
#energy----unit in joules
print(cons.calorie)
print(cons.eV)
print(cons.erg)
```

```python
#power---unit in watts
print(cons.hp)
```

```python
#force----unit in Newton
print(cons.kgf)
print(cons.lbf)
```

```python
#calculate of area of circle
radius=int(input("Enter the radius:"))
area=cons.pi*radius*radius           #area=pi*r*r
print("The area of the circle with radius is:",round(area,2),"sqm")
```

```python
#convert kilo meter into meter
kilo_meter=float(input("Enter the value:"))
meter=cons.kilo*kilo_meter             #meter=1000*km
print("After conversion:",meter,"m")
```

```python
#celcius to kelvin convert
celsius=float(input("Enter the tempareture in celsius:"))
kelvin=celsius+cons.zero_Celsius      #C=K-273.15
print("Celsius to Kelvin:",kelvin,"C")
```

```python
#import pandas library
import pandas as pd
import matplotlib.pyplot as plt
csv_file_path="C:/Users/Protiva/OneDrive/Desktop/Anudip_python/27 Excel/Cleaned_salesdata.csv"

#read the csv file
d_frame=pd.read_csv(csv_file_path)

#create the pivot table
pd.pivot_table(d_frame,index=["Region","Manager","SalesMan"],values=["Sale_amt"],aggfunc=["sum"])
```
